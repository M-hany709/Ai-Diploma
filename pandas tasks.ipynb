```python
#TASK2:how read random rows from data
```

```python
d=pd.read_excel(r"D:\menna\AI_COURSE\Dataset_4_sheets_200_rows.xlsx")
rows=np.random.choice(200,20,replace=False)
all_data=[]
for row in rows:
    all_data.append(d.iloc[[row]])
data=pd.concat(all_data,ignore_index=False) 
data=pd.DataFrame(data)
data
```

```python
df_existing = pd.read_excel(r"D:\menna\AI_COURSE\Dataset_4_sheets_200_rows.xlsx", sheet_name="Students")
df_new = pd.DataFrame({
    'Name': ['Aya', 'Omar'],
    'Age': [22, 24],
    "Gender":["female","male"],
    'GPA':[3,3.5],
    "Major":["teacher","doctor"]
})
data = pd.concat([df_existing, df_new], ignore_index=True)
data.to_excel(r"D:\menna\AI_COURSE\Dataset_4_sheets_200_rows.xlsx", sheet_name="Students", index=False)
data
```
