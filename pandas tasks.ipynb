```python
# 14. How do you concatenate DataFrames vertically using `concat()`?
```

```python
import pandas as pd
df1 = pd.DataFrame({
    'Name': ['Ali', 'Menna'],
    'Age': [23, 25]
})
df2 = pd.DataFrame({
    'Name': ['Omar', 'Sara'],
    'Age': [30, 28]
})
result = pd.concat([df1, df2], axis=0)
print(result)
```

```python
# 15. How do you merge DataFrames horizontally using `merge()`?
```

```python
import pandas as pd
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Ali', 'Menna', 'Omar']
})
df2 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Age': [23, 25, 30]
})
merged_df = pd.merge(df1, df2, on='ID')
print(merged_df)
```
